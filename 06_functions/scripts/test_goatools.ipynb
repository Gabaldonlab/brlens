```python
import os
import goatools
import pandas as pd
import math
```

```python
from goatools.obo_parser import GODag

if not os.path.exists('go-basic.obo'):
    !wget http://geneontology.org/ontology/go-basic.obo
go = GODag('go-basic.obo', optional_attrs=['relationship'])
```

```python
ipsdf = pd.read_csv('../outputs/YARLI_013.tsv', sep='\t', header=None)
```

```python
ipsdf
```

```python
odfl = list()
for i, item in enumerate(list(ipsdf[13])):
    if isinstance(item, str):
        if '|' in item:
            item = item.split('|')
            for subitem in item:
                try:
                    goterm = go[subitem]
                    odfd = dict()
                    odfd['id'] = ipsdf[0][i]
                    odfd['eval'] = ipsdf[8][i]
                    odfd[goterm.namespace] = goterm.name
                    odfl.append(odfd)
                except KeyError:
                    continue
        else:
            try:
                goterm = go[item]
                odfd = dict()
                odfd['id'] = ipsdf[0][i]
                odfd['eval'] = ipsdf[8][i]
                odfd[goterm.namespace] = goterm.name
                odfl.append(odfd)
            except KeyError:
                continue
```

```python
goterm.namespace
```

```python
def miau(l):
    s = ' / '.join(set([x for x in l if str(x) != 'nan']))
    return s
```

```python
odf = pd.DataFrame(odfl)
odf = odf.groupby('id').agg(miau)
```
